```python
fig = plt.figure()
year = ['1960', '1980', '2000', '2020']
male = [15.3, 24.2, 32.3, 40.2]
female = [12.3, 20.2, 27.8, 34.0]

plt.bar(year, male, align='center', color='red', alpha=0.5, label='Male')
plt.bar(year, female, align='center', color='blue', alpha=0.5, label='Female')
plt.title('Number of Never Married Adult in U.S.')
plt.xlabel('Year')
plt.ylabel('Number')
plt.legend()
plt.tight_layout()
```

```python
fig = plt.figure()
year = ['1960', '1980', '2000', '2020']
male = [15.3, 24.2, 32.3, 40.2]
female = [12.3, 20.2, 27.8, 34.0]
index = np.arange(len(year))
bwidth = 0.15
plt.bar(index-bwidth, male, width=bwidth, align='center', color='red', alpha=0.5, label='Male')
plt.bar(index, female, width=bwidth, align='center', color='blue', alpha=0.5, label='Female')
plt.title('Number of Never Married Adult in U.S.')
plt.xlabel('Year')
plt.ylabel('Number')
plt.xticks(index-bwidth/2, year)
plt.legend()
plt.tight_layout()
```

```python
data = [112, 100, 127, 120, 134, 118, 105, 110, 109, 112, 110, 118, 117, 116, 118, 122, 114, 114, 105, 109, 107, 112, 114, 115, 118, 117, 118, 122, 106, 110, 116, 108, 110, 121, 113, 120, 119, 111, 104, 111,
120, 113, 120, 117, 105, 110, 118, 112, 114, 114]

binwidth = 5


#boundaries = np.arange(min(data)-0.5, max(data)+binwidth, binwidth)
#hist, bin_edge = np.histogram(data, bins = boundaries)

mboundaries = np.hstack((min(boundaries)-5, boundaries.copy(), max(boundaries)+5))
hist, bin_edge = np.histogram(data, bins = mboundaries)

fig = plt.figure()
plt.hist(data, bins=bin_edge, color='skyblue', linewidth=1, edgecolor='black')
plt.yticks(np.arange(0, 21, 5))
plt.title('Maximum Temperature (F) in each City')
plt.xlabel('Frequency count')
plt.ylabel('Temperature (F)')
plt.ylim(0,20)
bin_centers = 0.5 * np.diff(bin_edge) + bin_edge[:-1]

for count, x in zip(hist,bin_centers):
    if count > 0:
        plt.annotate('{:d}'.format(count), xy=(x, (count-0.5)/20), xycoords=('data','axes fraction'), xytext=(0, 18), textcoords='offset points', va='top', ha='center') 
        
plt.plot(bin_centers, hist, '-', color='red')
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=1) # avg line

```

### Pie Graph

```python
fig = plt.figure(figsize=(8, 10))
snack = ['Potato Chips', 'Tortilla', 'Pretzels', 'Popcorn', 'Snack Nuts']
consumption = [11.2, 8.3, 4.3, 3.8, 2.5]
pcolors = ['gold', 'yellowgreen', 'aliceblue', 'lightskyblue', 'pink']
plt.pie(consumption, labels=snack, autopct='%1.1f%%', startangle=140, shadow=True, colors=pcolors) 
plt.title('The Amount of Snack Food Consuption (Million Pounds)')
plt.axis('equal')
plt.legend()
plt.tight_layout()
```
